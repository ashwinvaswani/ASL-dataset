```python
import cv2
import os
```

```python
current_dir = os.path.dirname(os.path.abspath('__file__'))
current_dir = current_dir + '/asl_alphabet_train/'
print(current_dir)
```

```python
for folder in sorted(os.listdir(current_dir)):
    #print(folder)
    new_dir = current_dir + folder + '/'
    print(new_dir)
    cnt = 1
    for filename in sorted(os.listdir(new_dir)):
        #print(filename[1:5]) 
        if cnt <= 200:
            os.remove(new_dir + filename)
            print(filename + "removed")           
            cnt += 1
```
